```python
ndvi_data = pd.read_csv('NDVI.csv')
gndvi_data = pd.read_csv('GNDVI.csv')
ndmi_date = pd.read_csv('NDMI.csv')
```

```python
ndvi_data = ndvi_data.rename(columns={
  'system:time_start': 'date_column',
  '0': 'ndvi'
})
gndvi_data = gndvi_data.rename(columns={
  'system:time_start': 'date_column',
  '0': 'gndvi'
})
ndmi_date = ndmi_date.rename(columns={
  'system:time_start': 'date_column',
  '0': 'ndmi'
})
```

```python
ndvi_data
```

```python
# prompt: Using dataframe ndvi_data: convert date format to yyyy-mm-dd

import datetime as dt
ndvi_data['date_column'] = ndvi_data['date_column'].apply(lambda x: dt.datetime.strptime(x, '%b %d, %Y').strftime('%Y-%m-%d'))
gndvi_data['date_column'] = gndvi_data['date_column'].apply(lambda x: dt.datetime.strptime(x, '%b %d, %Y').strftime('%Y-%m-%d'))
ndmi_date['date_column'] = ndmi_date['date_column'].apply(lambda x: dt.datetime.strptime(x, '%b %d, %Y').strftime('%Y-%m-%d'))
```

```python
if ndvi_data['date_column'].dtype == 'object':
    ndvi_data['date_column'] = pd.to_datetime(ndvi_data['date_column'])
elif xco2_agregate['date_column'].dtype == 'object':
    xco2_agregate['date_column'] = pd.to_datetime(xco2_agregate['date_column'])
```

```python
if gndvi_data['date_column'].dtype == 'object':
    gndvi_data['date_column'] = pd.to_datetime(gndvi_data['date_column'])
if ndmi_date['date_column'].dtype == 'object':
    ndmi_date['date_column'] = pd.to_datetime(ndmi_date['date_column'])
```

```python
import pandas as pd
merged_df = pd.merge(xco2_agregate, xco2_agregate, on='date_column', how='inner')
merged_df
```

```python
merged = pd.merge(merged_df, gndvi_data, on='date_column', how='inner')
merged
```

```python
final_merge = pd.merge(merged, ndmi_date, on='date_column', how='inner')
final_merge
```

```python
final_merge = final_merge.rename(columns={
  'date_column': 'date'
})
```

```python
final_merge
```

```python
final_merge = final_merge.drop_duplicates(subset='date', keep='first')
```

```python
final_merge_ndvi = final_merge[(final_merge['ndvi']> -0.01) &  (final_merge['ndvi']< 0.05)]
```

```python
# prompt: Plot Xco2 vs ndvi  and add a regression line to it

import altair as alt

# Define the data frame
df = final_merge

# Create the chart
chart = alt.Chart(df).mark_point().encode(
    x = 'ndmi:Q' ,
    y = alt.X('Xco2:Q', title = 'Xco2', scale=alt.Scale(domain=[400, 450])),
    color = alt.Color('Year:N', legend=None)
).properties(title='NDMI vs XCO2')

# Add a regression line
regression_line = chart.transform_regression('ndmi', 'Xco2').mark_line(color='red')

# Combine the chart and the regression line
final_chart = chart + regression_line

# Display the chart
final_chart
```
